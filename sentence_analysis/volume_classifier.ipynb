```python
def load_sentences(file):
    with open(file, encoding="utf-8") as f:
        lines = f.readlines()

    filtered = [line.strip() for line in lines]
    filtered = [line for line in filtered if line] # remove empty strings
    return filtered

def format_output(d):
    return f"{d['labels']} received scores {d['scores']}"

labels_1 = ["loud", "quiet"]
labels_2 = ["loud", "soft"]
labels_3 = ["noisy", "quiet"]
labels_4 = ["noisy", "soft"]
```

```python
keyword = "praise"

for year in range(1850, 2021, 10):
  print(f"Starting {year}")
  lines = load_sentences(f"drive/MyDrive/Colab Notebooks/{keyword}/{year}.txt")
  with open(f"drive/MyDrive/Colab Notebooks/{keyword}_results/{year}.txt", "w") as f:
    for i in range(0, len(lines), 3):
      f.write(lines[i] + "\n\n")
      f.write(lines[i + 1] + "\n\n")
      line = lines[i + 1]

      f.write(format_output(classifier(line, labels_1)) + "\n")
      f.write(format_output(classifier(line, labels_2)) + "\n")
      f.write(format_output(classifier(line, labels_3)) + "\n")
      f.write(format_output(classifier(line, labels_4)) + "\n\n")
print("Finished!")
```

```python
"""
PARAGRAPH ANALYZER
"""

def get_paragraphs(file_name):
    """
        TODO:  post 1941, files have no header line
    """

    paragraphs = []

    with open(file_name, encoding="utf-8") as f:
        # next(f)
        # next(f) # get rid of blank line
        cur_paragraph = ""
        for line in f:
            val = line.strip()
            if val:
                cur_paragraph += f"{val} "
            else:
                if cur_paragraph:
                    paragraphs.append(cur_paragraph)
                cur_paragraph = ""
        if cur_paragraph:
            paragraphs.append(cur_paragraph) # just in case

    return paragraphs

def paragraph_output(d, num_words):
    return f"{d['labels']} {d['scores']} {num_words}"

labels_1 = ["loud", "quiet"]
labels_2 = ["loud", "soft"]
labels_3 = ["noisy", "quiet"]
labels_4 = ["noisy", "soft"]
```
