```python
def load_sentences(file):
    with open(file, encoding="utf-8") as f:
        lines = f.readlines()

    filtered = [line.strip() for line in lines]
    filtered = [line for line in filtered if line] # remove empty strings
    return filtered

def format_output(d):
    return f"{d[0]['label']} received score {d[0]['score']}"

keyword = "praise"

# TEST
# print(format_output(sentiment_analysis("I love this!")))

test_lines = ["We'll sing and we'll shout with the armies of heaven",
              # "Bella tiptoed through the corridor",
              "Give heed to the whisperings of the Holy Spirit",
              "I refer to pronounced irreverence in our church services, with the noise, laughter and confusion that often accompanies such a condition"
              # "Bev roared like a lion upon receiving the news"
              ]

for line in test_lines:
  print(line)
  print(format_output(sentiment_analysis(line)))
  print()


```

```python
"""
PARAGRAPH ANALYZER
"""

def get_paragraphs(file_name):
    paragraphs = []

    with open(file_name, encoding="utf-8") as f:
        next(f)
        next(f) # get rid of blank line
        cur_paragraph = ""
        for line in f:
            val = line.strip()
            if val:
                cur_paragraph += f"{val} "
            else:
                if cur_paragraph:
                    paragraphs.append(cur_paragraph)
                cur_paragraph = ""
        if cur_paragraph:
            paragraphs.append(cur_paragraph) # just in case

    return paragraphs

def paragraph_output(d, num_words):
    return f"{d[0]['label']} {d[0]['score']} {num_words}"
```

```python
file_name = f"drive/MyDrive/Colab Notebooks/jfs/1890/1897_o.txt"

paragraphs = get_paragraphs(file_name)
for p in paragraphs:
  num_words = len(p.split())
  print(paragraph_output(sentiment_analysis(p), num_words))
  print()
```
